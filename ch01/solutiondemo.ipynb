# Exercise 2.9: Expected Output Changes and Permanent Income

Equation  
$$
y_t - y^p_t = \sum_{j=1}^{\infty} \frac{1}{(1+r)^j} E_t \Delta y_{t+j}
$$  
expresses the difference between current and permanent income, $y_t - y^p_t$, as the present discounted value of expected future changes in the endowment.  

**Task:**  
- Present a step-by-step derivation starting from:  
  $$
  y^p_t = \frac{r}{1+r} \sum_{j=0}^{\infty} \frac{E_t y_{t+j}}{(1+r)^j}
  $$
  and  
  $$
  \Delta y_{t+j} = y_{t+j} - y_{t+j-1}
  $$
- Comment on the cyclical properties of $y_t - y^p_t$ depending on whether the **level** or the **change** of $y_t$ follows an AR(1) process.

## Step-by-Step Derivation

Multiply  
$$
y^p_t = \frac{r}{1+r} \sum_{j=0}^{\infty} \frac{E_t y_{t+j}}{(1+r)^j}
$$  
by $\frac{1+r}{r}$:  
$$
\frac{1+r}{r} y^p_t = \sum_{j=0}^{\infty} \frac{E_t y_{t+j}}{(1+r)^j}
$$

Split the sum:  
$$
\frac{1+r}{r} y^p_t = y_t + \sum_{j=1}^{\infty} \frac{E_t y_{t+j}}{(1+r)^j}
$$

Subtract $y^p_t$ from both sides:  
$$
\frac{1}{r} y^p_t = y_t - y^p_t + \sum_{j=1}^{\infty} \frac{E_t y_{t+j}}{(1+r)^j}
$$

Rearrange:  
$$
y_t - y^p_t = - \sum_{j=1}^{\infty} \frac{E_t y_{t+j}}{(1+r)^j} + \frac{1}{r} y^p_t
$$

Divide the original $y^p_t$ expression by $r$:  
$$
\frac{1}{r} y^p_t = \sum_{j=0}^{\infty} \frac{E_t y_{t+j}}{(1+r)^{j+1}}
$$

Substitute back:  
$$
y_t - y^p_t = - \sum_{j=1}^{\infty} \frac{E_t y_{t+j}}{(1+r)^j} + \sum_{j=0}^{\infty} \frac{E_t y_{t+j}}{(1+r)^{j+1}}
$$

Simplify:  
$$
y_t - y^p_t = - \sum_{j=1}^{\infty} \frac{1}{(1+r)^j} (E_t y_{t+j} - E_t y_{t+j-1})
$$

Using $\Delta y_{t+j} = y_{t+j} - y_{t+j-1}$:  
$$
y_t - y^p_t = \sum_{j=1}^{\infty} \frac{1}{(1+r)^j} E_t \Delta y_{t+j}
$$


## Cyclical Properties

### Case 1: $y_t$ is AR(1)
From:  
$$
y^p_t = \frac{r}{1+r-\rho} y_t
$$  
we have:  
$$
y_t - y^p_t = \frac{1-\rho}{1+r-\rho} y_t
$$  
Since $\frac{1-\rho}{1+r-\rho} > 0$,  
$y_t - y^p_t$ is **perfectly procyclical** ($corr(y_t - y^p_t, y_t) = 1$).

### Case 2: $\Delta y_t$ is AR(1)
$$
E_t \Delta y_{t+j} = \rho^j \Delta y_t
$$  
so:  
$$
y_t - y^p_t = -\frac{\rho}{1+r-\rho} \Delta y_t
$$  
Since $\frac{-\rho}{1+r-\rho} < 0$,  
$y_t - y^p_t$ is **perfectly countercyclical** ($corr(y_t - y^p_t, \Delta y_t) = -1$).
